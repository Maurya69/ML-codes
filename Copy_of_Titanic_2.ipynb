```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/train.csv")
copy_df=df
```

```python
df
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_train.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
transformer=ColumnTransformer(
    [
        ('knn',KNNImputer(),[2]),
        ('si',SimpleImputer(strategy='most_frequent'),[6]),
    ],remainder='passthrough')
```

```python
X_train_transformed=transformer.fit_transform(X_train)
X_test_transformed=transformer.transform(X_test)
```

```python
trans=ColumnTransformer(
    [
        ('ohe',OneHotEncoder(sparse=False),[1]),
        ('ohe1',OneHotEncoder(sparse=False),[3])
    ],remainder='passthrough'
)
```

```python
d=pd.DataFrame(X_train_transformed)
c=pd.DataFrame(X_test_transformed)
```

```python
X_tt=trans.fit_transform(d)
X_t=trans.transform(c)
```

```python
dd=pd.DataFrame(X_tt)
cc=pd.DataFrame(X_t)
```

```python
dd
```

```python
cc
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_tt=ss.fit_transform(X_tt)
X_t=ss.transform(X_t)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(max_depth=5,n_estimators=500)
```

```python
clf.fit(X_tt,y_train)
y_pred = clf.predict(X_t)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
